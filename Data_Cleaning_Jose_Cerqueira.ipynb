```python

gender_types = insurance_df["gender"].unique()
gender_types
```

```python

insurance_df["gender"] = insurance_df["gender"].replace("Femal", "F")
insurance_df["gender"] = insurance_df["gender"].replace("female", "F")
insurance_df["gender"] = insurance_df["gender"].replace("Male", "M")

gender_types = insurance_df["gender"].unique()
gender_types
```

```python
state_types = insurance_df["state"].unique()
state_types
```

```python
insurance_df["state"] = insurance_df["state"].replace("cali", "California")
insurance_df["state"] = insurance_df["state"].replace("AZ", "Arizona")
insurance_df["state"] = insurance_df["state"].replace("WA", "Washington")

state_types = insurance_df["state"].unique()
state_types
```

```python
educ_types = insurance_df["education"].unique()
educ_types
```

```python
insurance_df["education"] = insurance_df["education"].replace("Bachelors", "Bachelor")

educ_types = insurance_df["education"].unique()
educ_types
```

```python
customer_life_type = insurance_df["cust_lifetime_value"].unique()
customer_life_type
```

```python
insurance_df["cust_lifetime_value"] = insurance_df["cust_lifetime_value"].str.replace("%", " ")
customer_life_type = insurance_df["cust_lifetime_value"].unique()
customer_life_type
```

```python
vehicles_class = insurance_df["vehicle_class"].unique()
vehicles_class 
```

```python
insurance_df["vehicle_class"] = insurance_df["vehicle_class"].replace("Luxury SUV", "Luxury")
insurance_df["vehicle_class"] = insurance_df["vehicle_class"].replace("Luxury Car", "Luxury")
insurance_df["vehicle_class"] = insurance_df["vehicle_class"].replace("Sports Car", "Luxury")

vehicles_class = insurance_df["vehicle_class"].unique()
vehicles_class 
```
